````python
import matplotlib.pyplot as plt
import numpy as np

# evenly sampled time at 200ms intervals
t = np.arange(0., 5., 0.2)

# red dashes, blue squares and green triangles
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
````

numpy 모듈을 임포트하고 np라는 이름으로 부르겠다는 것을 확인하실 수 있습니다. 즉 numpy 모듈의 기능을 사용하여 코드가 작성될 것이라는 것을 짐작할 수 있습니다. 

numpy 모듈을 사용하지 않고 위 코드셀의 코드와 동일한 결과를 산출하는 코드를 만들어 보겠습니다.

````python
import matplotlib.pyplot as plt

t  = []
p2 = []
p3 = []

for i in range(0, 50, 2):
  t.append( i/10 )
  p2.append( (i/10)**2 )
  p3.append( (i/10)**3 )

plt.plot(t, t, 'r--', t, p2, 'bs', t, p3, 'g^')
plt.show()
````

numpy 모듈을 사용해서 코딩한 것과 numpy 모듈을 사용하지 않고 작성한 코드 중 어느 코드가 더 간결하다고 생각되십니까? numpy를 지금 처음 접하신다면 제 질문에 "글세~"라고 대답하실 수 있습니다. 

~그렇다면 이번에는 실행 속도를 비교해 보겠습니다. numpy 모듈을 활용하지 않고 작성된 코드와 numpy 모듈을 활용한 코드의 실행 속도 비교를 위해 아래 코드를 봐주세요.~

아래 코드도 속도 비교를 위해 적절한지 다시 생각해 봐야할 것 같고 무엇보다 지금 이 이야기를 할 때가 아닌 것 같습니다.

````python
import numpy as np
num = 1000

def WithoutNumpy(n):
  a = list( range(n) )
  b = list( range(n) )
  c = []

  for i in range(n):
    a[i] = 2 * a[i]
    b[i] = b[i]**2
    c.append(a[i]+b[i])
  return c

def WithNumpy(n):
  a = 2 * np.arange(n)
  b = np.arange(n) ** 2
  return a + b

print("WithOutNumpy()")
%timeit WithoutNumpy(num)

print()

print("WithNumpy()")
%timeit WithNumpy(num)
````

원주율, pi(π)의 값을 소주점 몇 째자리까지 기억하십니까? 3.14....?

````python
#원의 둘레
pi = 3.14
r = 10 #[m]

print('반지름이 {}[m]인 원의 둘레는 {}[m]입니다'.format(r, 2*pi*r))
print('즉 원의 둘레는 {:.1f}[cm]입니다'.format(2*pi*r * 100))     # 1m = 100cm
````

````python
import numpy as np
r = 10 #[m]
print('반지름이 {}[m]인 원의 둘레는 {}[m]입니다'.format(r, 2*np.pi*r))
print('즉 원의 둘레는 {:.1f}[cm]입니다'.format(2*np.pi*r * 100))  # 1m = 100cm
````

위 두 코드셀의 출력 결과가 다릅니다. 원주율, pi(π) 값이 잘 생각나지 않는다고 소수점 이하 두 째자리까지만 썼다가는 큰 일이 날 수도 있겠습니다. 원주율이 얼마인지 기억하고 있어야 하는 것이 아니라 numpy 모듈을 이용하여 위 코드에서와 같이 ```np.pi```라고 표현하면 되겠습니다.

다음으로 무작위(無作爲)로 값을 생성하는 랜덤(random) 함수를 살펴보겠습니다. UNIT 6에서 한번 사용한 적이 있습니다. Python에서 제공하는 random 모듈에는 많은  기능(함수)가 있습니다만 지금은 ```randint()``` 함수만 살펴보겠습니다.

````python
import matplotlib.pyplot as plt
import random
dice = []

random.seed(2023)     #이 코드를 실행하는 사람들 여러 번 코드를 실행 해도 모두 동일한 결과를 얻기 위해.

for i in range(10):
  dice.append( random.randint(1, 6) )

plt.hist(dice, bins=6)
plt.show()
````

```random.randint()```함수가 반환하는 데이터의 형(타입, type)은 무엇입니까?

````python
n = random.randint(1,6)
print( n )
print( type(n) )
````
