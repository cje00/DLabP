위 두 코드셀의 실행 결과에서 알 수 있듯이 먼저 `range(0)`는 일정 범위를 만들지 못했다는 것을 확인했습니다.   

그리고 `for in range(month-1)`은 `month=1`일 때 결국 `for in range(0)`과 같습니다. 결국 `for in` 구조에서 `in`뒤에 이터러블한 데이터가 와서 문법 에러가 나지는 않지만 반복 해서 꺼낼 데이터가 없기 때문에 `for`에 의한 반복 동작이 일어나지 않습니다. 만약 `for`문에 의한 반복 동작이 일어났다면 최소한 한번은 `print('in for')`가 실행되었어야 합니다. 따라서 `total+=month_list[i]` 문장도 실행될 수 없습니다. 

다만 `for`문 밖 `total+=day` 문장부터 실행되어 `total-1`의  값 1이 출력된 것입니다. "1월 2일은 새해 첫날(1월 1일)을 기준으로 1일이 지난 날입니다. 


````python
func_a(1,3)
````

````python
month = 1
day = 3
month_list = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
total = 0;

for i in range(month - 1):
  total += month_list[i]    # 월별 날자
  print('in for')

total += day               # 날자 
print(total-1)
````

func_a(1,3)을 실행하면 1월 3일은 새해 첫날(1월 1일)을 기준으로 2일이 지난 날입니다.

1월 2일과 1월 3일 몇일 떨어져 있냐는 물음에 "2일지난 날에서 1일 지난 날을 빼니 1일의 차이가 생기네요.

````python
def func_a(month, day):
    month_list = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
    total = sum(month_list[:month-1]) + day
    return total - 1
        
def solution(start_month, start_day, end_month, end_day):
    start_total = func_a(start_month, start_day)
    end_total = func_a(end_month, end_day)
    return end_total - start_total

solution(1, 2, 1, 3)
````

단, 몇 배 더 많은지 구할 때는 소수 부분은 버리고 정수 부분만 구하면 됩니다.

자연수가 들어있는 리스트 arr가 매개변수로 주어질 때, 가장 많이 등장하는 숫자가 가장 적게 등장하는 숫자보다 몇 배 더 많은지 return 하도록 solution 함수를 작성하려 합니다. 위 구조를 참고하여 코드가 올바르게 동작할 수 있도록 빈칸에 주어진 func_a, func_b, func_c 함수와 매개변수를 알맞게 채워주세요.

---
##### 매개변수 설명
자연수가 들어있는 리스트 arr가 solution 함수의 매개변수로 주어집니다.
* arr의 길이는 3 이상 1,000 이하입니다.
* arr에는 1 이상 1,000이하의 자연수가 들어있습니다.

---
##### return 값 설명
리스트에서 가장 많이 등장하는 숫자가 가장 적게 등장하는 숫자보다 몇 배 이상 많은지 return 해주세요.

* 가장 많이 들어있는 수의 개수와 가장 적게 들어있는 수의 개수가 같은 경우에는 1을 return 합니다.

---
##### 예시

| arr                   | return |
|-----------------------|--------|
| [1,2,3,3,1,3,3,2,3,2] | 2      |

##### 예시 설명
리스트에 1이 2개, 2가 3개, 3이 5개 들어있습니다.

* 가장 적게 들어있는 숫자 : 1 (2개)
* 가장 많이 들어있는 숫자 : 3 (5개)

3이 1보다 2.5배 많이 들어있으며, 소수 부분을 버리고 2를 return 하면 됩니다.

#### 여기는 참고만 하세요.

<font color=blue> 자연수? 자연수는 양의 정수! 일단 위 1, 2, 3, 4단계를 적용하여 스스로 코드를 한번 만들어보세요.</font>



````python
# 단계 1 -리스트에 들어있는 각 자연수의 개수를 셉니다.
lst = [1, 2, 0, -1.2, -4, 3.14, 5, 5, 6, 6, 6, 7, 7, 7, 7]
mlst = []
for value in lst:
  if value > 0 and isinstance(value, int):
    mlst.append(value)
print(mlst)

unique = set(mlst)
result = {}

for value in unique:
  result[value] = lst.count(value)

print(result)
````

````python
# 단계 1 -리스트에 들어있는 각 자연수의 개수를 셉니다.
from collections import Counter

lst = [1, 2, 0, -1.2, -4, 3.14, 5, 5, 6, 6, 6, 7, 7, 7, 7]
mlst = []

for value in lst:
  if value > 0 and isinstance(value, int):
    mlst.append(value)

print(mlst)

counter = Counter(mlst)
print(counter)
d_counter = dict(counter)
print(d_counter)
````

````python
# 단계 2 - 가장 많이 등장하는 수의 개수를 구합니다.

# value(값)을 기준으로 정령
ss = sorted(d_counter.items(), key=lambda x:x[1])
print(ss)
print(ss[-1])

# key을 기준으로 정령
#sorted(d_counter.items(), key=lambda x:x[0])
````

````python
# 단계 3 - 가장 적게 등장하는 수의 개수를 구합니다.
print(ss[0])
````

````python
# 단계 4 - 가장 많이 등장하는 수가 가장 적게 등장하는 수보다 몇 배 더 많은지 구합니다.
print(ss[-1][1])
print(ss[0][1])

print(ss[-1][1] /ss[0][1])
````
