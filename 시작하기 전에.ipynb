# 저자가 이 책에 대해서 한 말
최소한의 문법과 입출력을 위해 최소한의 라이브러리를 사용해서 여러분만의 프로젝트를 만들 수 있도록 도와줌



## 누구를 위한 책인가요?
파이썬 기초까지는 배웠지만 그 다음 단계로 무엇을 할지 궁금했던 사람, 파이썬으로 무언가 나만의 프로젝트를 해보고 싶었던 사람을 두 대상으로 합니다.

## 파이썬을 몰라도 괜찮나요?
입출력이나 for, if ,list 정도의 기초 문법만 알면 누구가 이 책을 읽을 수 있습니다. 하지만 컴퓨터 프로그램을 한 번도 접해본 적이 없는 사람이라면 부록의 '파이썬 병아리반' 내용을 먼저 볼 것을 추천합니다.
